- In Lahman's Baseball Database table `people`, there is information about people's `birthCountry` and `deathCountry`.<br><br>

- There are countries in which at least one person was born but in which no person has died.<br><br>

- Write a query that produces a table of the form:
    - `birthCountry`
    - `no_of_births`, which is the total number of births in the country.<br><br>

- The table contains all rows for countries in which there with births but no deaths.<br><br>

- You may create indexes to improve performance.

<u>Answer</u>


````python
%%sql
SELECT
    birthCountry AS birthCountry,
    COUNT(*) AS no_of_births
FROM
    people
WHERE
    birthCountry IS NOT NULL
    AND birthCountry NOT IN (
        SELECT DISTINCT deathCountry
        FROM people
        WHERE deathCountry IS NOT NULL
    )
GROUP BY
    birthCountry;
````

    - ```win_percentage``` is the winning percentage.

    - All number become floating point. Your answer uses SQL and the values should be integers except for ```win_percentage.```

<u>Answer</u>


````python
%%sql
WITH BattingCriteria AS (
    SELECT
        p.playerid,
        p.nameLast,
        p.nameFirst,
        'Batter' AS great_because,
        (
            (SUM(b.H) + SUM(b.BB)) / (SUM(b.AB) + SUM(b.BB))
        ) AS obp,
        (
            (
                (SUM(b.H) + SUM(b."2B") + 2 * SUM(b."3B") + 3 * SUM(b.HR))
                / SUM(b.AB)
            )
        ) AS slg,
        (
            (
                (SUM(b.H) + SUM(b.BB)) / (SUM(b.AB) + SUM(b.BB))
                + (
                    (SUM(b.H) + SUM(b."2B") + 2 * SUM(b."3B") + 3 * SUM(b.HR))
                    / SUM(b.AB)
                )
            )
        ) AS obp_slg,
        SUM(b.AB) AS total_abs
    FROM
        people AS p
    LEFT JOIN
        batting AS b
    ON
        p.playerid = b.playerid
    GROUP BY
        p.playerid
    HAVING
        SUM(b.AB) >= 3000
        AND obp_slg >= 1.000
),
PitchingCriteria AS (
    SELECT
        p.playerid,
        p.nameLast,
        p.nameFirst,
        'Pitcher' AS great_because,
        NULL AS obp,
        NULL AS slg,
        NULL AS obp_slg,
        NULL AS total_abs,
        SUM(p.W) AS total_w,
        SUM(p.L) AS total_l,
        SUM(p.W + p.L) AS total_d,
        (SUM(p.W) / SUM(p.W + p.L)) AS win_percentage
    FROM
        people AS p
    LEFT JOIN
        pitching AS p
    ON
        p.playerid = p.playerid
    GROUP BY
        p.playerid
    HAVING
        (SUM(p.W + p.L) >= 200 AND win_percentage >= 0.70) OR SUM(p.W) >= 300
)
SELECT
    playerid,
    nameLast,
    nameFirst,
    great_because,
    slg,
    obp,
    obp_slg,
    total_abs,
    total_w,
    total_l,
    total_d,
    win_percentage
FROM
    BattingCriteria
UNION
SELECT
    playerid,
    nameLast,
    nameFirst,
    great_because,
    slg,
    obp,
    obp_slg,
    total_abs,
    total_w,
    total_l,
    total_d,
    win_percentage
FROM
    PitchingCriteria;

````
